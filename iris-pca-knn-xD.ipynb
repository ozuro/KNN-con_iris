```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
from mpl_toolkits.mplot3d import Axes3D
from pylab import rcParams
import seaborn as sb

import scipy
from scipy.stats.stats import pearsonr
%matplotlib inline 
```

```python
data_set = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
```

```python
# load dataset into Pandas DataFrame
rcParams['figure.figsize']=5,4
sb.set_style('whitegrid')
df = pd.read_csv(data_set, names=['sepal length','sepal width','petal length','petal width','target'])
#rint(df)
sb.pairplot(df)

```

```python
#df=df.drop(['sepal width'], axis =1)
SepalLength= df['petal length']
SepalWidth= df['petal width']

pearsonr_coeficiente, p_valor = pearsonr(SepalLength,SepalWidth)
print('r de person: ', pearsonr_coeficiente)
print('probabilidad de r : ', p_valor )

```

```python
#estandarizacion de los datos
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
features = ['sepal length', 'sepal width', 'petal length', 'petal width']
# SeparaR las caracteristicas
x = df.loc[:, features].values
#print(x)
# Separar el target
y = df.loc[:,['target']].values
# Estandarizando las caracteristicas

#scaler = MinMaxScaler()
#x = scaler.fit_transform(x)
#print(x)
new_y =[]
for i in y:
    new_y.append( i[0] )
y = new_y
print (y)
```

```python
#aplicacion del algoritmo PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2', 'principal component 3'])
print(principalDf)
#df=df.drop(['sepal width'], axis =1)
```

```python
# finalDf = pd.concat(principalDf)
final = np.array (principalDf)
print(final)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (x,y, test_size = 0.33)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier( n_neighbors = 3 ).fit(x_train,y_train)

# display (clf.kneighbors_graph(x_train).toarray())
print (clf)
```

```python
from sklearn.metrics import accuracy_score
print ("accuracy:")
print ( accuracy_score ( y_test, clf.predict(x_test) ) )
print ("Clasificacion:")
#0 = setosa
#1 =versicolor
#2 = virginica


datos = [[5.0, 0.0, 5.0 , 1.5], [5.0,0.0,2.0,1.5], [6.0,0.0,5.0,1.5],[6.0,0.0,5.0,2.0]]
print( clf.predict(datos) )
```
